```python
import numpy as np
import trimesh
import geopandas as gpd
import pandas as pd
from shapely import geometry

from boreholes_helper import Boreholes

pd.options.mode.chained_assignment = None  # default='warn'

a = Boreholes()

# set collar
a.setCollar(
    path='./BAKAN_COLLAR.CSV',
    x='UTM_EAST',
    y='UTM_NORTH',
    z='RL',
    depth='TDEPTH',
    keys=['BHID'],
    id='BHID',
)
# read collar and apply srs
a.readCollar()
a.applyCollarConfigMap()
a.createCollarPointGeometry()
a.updateCollarCrs(32649)

# create metric geometry
df = a.getDataFrame('collar')
df = Boreholes.changeCrs(df, srid=32649)
df['geometryMetric'] = df['geometry']
df = Boreholes.changeCrs(df, srid=4326)
a.setDataFrame('collar', df)

a.setCollar(srid=4326)
a.createCollarLineSegment(key='geometryMetric')
# a.viewCollarLine().show()

# set survey
a.setSurvey(
    path='./BAKAN_SURVEY.CSV',
    depth='AT',
    dip='DM_DIP',
    azimuth='UTM_BRG',
    keys=['BHID'],
    id='BHID',
)
a.readSurvey()
a.applySurveyConfigMap()

# set interval
a.setInterval(
    path='./BAKAN_ASSAY.CSV',
    z0='FROM',
    z1='TO',
    keys=['BHID']
)
a.readInterval()
a.applyIntervalConfigMap()
```
